### Required Assignment 5.1: Will the Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaurant near where you are driving. Would you accept that coupon and take a short detour to the restaurant? Would you accept the coupon but use it on a subsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaurant? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\$20 - $50).

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. You will publish your findings in a public facing github repository as your first portfolio piece.





### Data Description
Keep in mind that these values mentioned below are average values.

The attributes of this data set include:
1. User attributes
    -  Gender: male, female
    -  Age: below 21, 21 to 25, 26 to 30, etc.
    -  Marital Status: single, married partner, unmarried partner, or widowed
    -  Number of children: 0, 1, or more than 1
    -  Education: high school, bachelors degree, associates degree, or graduate degree
    -  Occupation: architecture & engineering, business & financial, etc.
    -  Annual income: less than \\$12500, \\$12500 - \\$24999, \\$25000 - \\$37499, etc.
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she buys takeaway food: 0, less than 1, 1 to 3, 4 to 8 or greater
    than 8
    -  Number of times that he/she goes to a coffee house: 0, less than 1, 1 to 3, 4 to 8 or
    greater than 8
    -  Number of times that he/she eats at a restaurant with average expense less than \\$20 per
    person: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    

2. Contextual attributes
    - Driving destination: home, work, or no urgent destination
    - Location of user, coupon and destination: we provide a map to show the geographical
    location of the user, destination, and the venue, and we mark the distance between each
    two places with time of driving. The user can see whether the venue is in the same
    direction as the destination.
    - Weather: sunny, rainy, or snowy
    - Temperature: 30F, 55F, or 80F
    - Time: 10AM, 2PM, or 6PM
    - Passenger: alone, partner, kid(s), or friend(s)


3. Coupon attributes
    - time before it expires: 2 hours or one day

### Problems

Use the prompts below to get started with your data analysis.  

1. Read in the `coupons.csv` file.




```python
data = pd.read_csv('data/coupons.csv')
```

```python
data.head()
```

2. Investigate the dataset for missing or problematic data.

```python
# Print row count
row_count = data.shape[0]
print(row_count)

# Print missing values
missing_values = data.isnull().sum()
print(missing_values)
```

3. Decide what to do about your missing data -- drop, replace, other...

```python
#Most values have the car column missing. Lets investigate further.
data["car"].value_counts()

# Since this column is barely used. Lets remove it from our dataset
data = data.drop(columns=['car'])

# check that its gone
print(data.columns) 
```

```python
#There are lots of rows where a visit frequency column is missing
missing_any_visit_frequency_column = data[data["Bar"].isnull() | data["CoffeeHouse"].isnull() | data["CarryAway"].isnull() | data["RestaurantLessThan20"].isnull() & data["Restaurant20To50"].isnull()].shape[0]

# Removing all of these values would only remove 3% of our dataset
missing_data_percentage = missing_any_visit_frequency_column / data.shape[0]
print(missing_data_percentage)

# so lets remove them all
data = data.dropna(
    subset=["Bar", "CoffeeHouse", "CarryAway", "RestaurantLessThan20", "Restaurant20To50"],
    how='any'
)

# And check that they are gone
data[data["Bar"].isnull() | data["CoffeeHouse"].isnull() | data["CarryAway"].isnull() | data["RestaurantLessThan20"].isnull() | data["Restaurant20To50"].isnull()].shape[0]
```

```python
# We no longer have missing values
missing_values = data.isnull().sum()
print(missing_values)
```

4. What proportion of the total observations chose to accept the coupon?



```python
accepted_coupon = data[data["Y"] == 1]
rejected_coupon = data[data["Y"] == 0]

# Over 50% accepted the coupon
proportion_accepted_coupon = accepted_coupon.shape[0] / data.shape[0]
print(proportion_accepted_coupon)

# Less than 50% rejected the coupon
proportion_rejected_coupon = rejected_coupon.shape[0] / data.shape[0]
print(proportion_rejected_coupon)
```

5. Use a bar plot to visualize the `coupon` column.

```python
#Lets explore the coupon column
data['coupon'].value_counts()

# Lets set the data column to a string
data['coupon'] = data['coupon'].astype(str)

# Lets plot the coupon column
sns.countplot(x='coupon', data=data)
plt.title('Coupon Count')
plt.xlabel('Coupon Type')
plt.ylabel('Coupon Count')
plt.xticks(rotation=45)
plt.show()
```

6. Use a histogram to visualize the temperature column.

```python
#Lets explore the temperature column
data['temperature'].value_counts()

plt.hist(data['temperature'], color='blue')
plt.title('Temperature at which Coupon was given')
plt.xlabel('Temperature')
plt.ylabel('Coupon Count')
plt.show()
```

**Investigating the Bar Coupons**

Now, we will lead you through an exploration of just the bar related coupons.  

1. Create a new `DataFrame` that contains just the bar coupons.


```python
bar_coupons = data[data["coupon"] == "Bar"]
# copy the dataset
bar_coupons = bar_coupons.copy()
bar_coupons.head()
```

2. What proportion of bar coupons were accepted?


```python
proportion_accepted_coupons = bar_coupons["Y"].mean()
print(proportion_accepted_coupons)
```

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
# Get 3 or less acceptance rate
three_or_fewer = bar_coupons[bar_coupons['Bar'].isin(['never', 'less1', '1~3'])]
three_or_fewer_acceptance_rate = three_or_fewer["Y"].mean()

# Get 4 or more acceptance rate
four_or_more = bar_coupons[bar_coupons['Bar'].isin(['4~8', 'gt8'])]
four_or_more_acceptance_rate = four_or_more["Y"].mean()


# Create a bar plot comparing both values
plt.bar(["Three or Fewer", "Four or More"], [three_or_fewer_acceptance_rate, four_or_more_acceptance_rate])
plt.title('Coupon Acceptance Rate by Bar Visits')
plt.ylabel('Coupon Acceptance Rate')
plt.xlabel('Visits')
plt.ylim(0, 1)  # Rates range between 0 and 1
plt.show()
```

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
# First convert age to an int and set ages 50+ to 50 and under 21 to 20
def convert_age_to_int(age):
    if age == "50plus":
        return 50
    elif age == "below21":
        return 20
    return int(age)

# Apply conversion function
bar_coupons['age'] = bar_coupons['age'].apply(convert_age_to_int)

# Get the drivers over 25 and who visit a bar more than once a month
over_25_visited_bar = bar_coupons[(bar_coupons['Bar'].isin(['1~3', '4~8', 'gt8'])) & (bar_coupons["age"] > 25)]
over_25_visited_bar_acceptance_rate = over_25_visited_bar["Y"].mean()

# get the rest of the drivers
not_in_over_25 = bar_coupons[~bar_coupons.index.isin(over_25_visited_bar.index)]
not_in_over_25_acceptance_rate = not_in_over_25["Y"].mean()

# Create a bar plot comparing both values
plt.bar(["age 26+, bar visits/month 1+", "others"], [over_25_visited_bar_acceptance_rate, not_in_over_25_acceptance_rate])
plt.title('Coupon Acceptance Rate by Age > 25 and visits bars 1+ times per month')
plt.ylabel('Coupon Acceptance Rate')
plt.xlabel('Visits')
plt.ylim(0, 1)
plt.show()
```

5. Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry.


```python
# Get the drivers who visit a bar more than once a month, had a non-child passenger, and weren't in farming, fishing or forestry
group = bar_coupons[(bar_coupons['Bar'].isin(['1~3', '4~8', 'gt8'])) & (~bar_coupons["passanger"].isin(["Alone", "Kid(s)"])) & (bar_coupons["occupation"] != "Farming Fishing & Forestry")]
group_acceptance_rate = group["Y"].mean()

# get the rest of the drivers
not_in_group = bar_coupons[~bar_coupons.index.isin(over_25_visited_bar.index)]
not_in_group_acceptance_rate = not_in_over_25["Y"].mean()

# Create a bar plot comparing both values
plt.bar(["Group", "Not Group"], [group_acceptance_rate, not_in_group_acceptance_rate])
plt.title('Coupon Acceptance Rate by Age > 25 and visits bars 1+ times per month')
plt.ylabel('Coupon Acceptance Rate')
plt.xlabel('Visits')
plt.ylim(0, 1)
plt.show()
```

6. Compare the acceptance rates between those drivers who:

- go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*
- go to bars more than once a month and are under the age of 30 *OR*
- go to cheap restaurants more than 4 times a month and income is less than 50K.



```python

# go to bars more than once a month, had passengers that were not a kid, and were not widowed OR
group1 = bar_coupons[(bar_coupons['Bar'].isin(['1~3', '4~8', 'gt8'])) & (~bar_coupons["passanger"].isin(["Alone", "Kid(s)"])) & (bar_coupons["maritalStatus"] != "Widowed")]
group1_acceptance_rate = group1["Y"].mean()
# print(group1)

# go to bars more than once a month and are under the age of 30 OR
group2 = bar_coupons[(bar_coupons['Bar'].isin(['1~3', '4~8', 'gt8'])) & (bar_coupons["age"] < 30)]
group2_acceptance_rate = group2["Y"].mean()
# print(group2)

# go to cheap restaurants more than 4 times a month and income is less than 50K.
group3 = bar_coupons[(bar_coupons["RestaurantLessThan20"].isin(['4~8', 'gt8'])) & (bar_coupons["income"].isin(["Less than $12500", "$12500 - $24999", "$25000 - $37499", "$37500 - $49999"]))]
group3_acceptance_rate = group3["Y"].mean()
# print(group3)

# All
all_acceptance_rate = bar_coupons["Y"].mean()

# # Create a bar plot comparing all values
plt.bar(["Group 1", "Group 2", "Group 3", "All"], [group1_acceptance_rate, group2_acceptance_rate, group3_acceptance_rate, all_acceptance_rate])
plt.title('Coupon Acceptance Rate by Group')
plt.ylabel('Coupon Acceptance Rate')
plt.ylim(0, 1)
plt.show()
```

7.  Based on these observations, what do you hypothesize about drivers who accepted the bar coupons?

```python
# Based on the above observations, it seems that the greatest indicator of whether a driver will accept a bar coupon or not is
# whether or not they attend bars regularly. If they visit bars at least once per month, they are likely to accept the coupon 
# becuase they are likely going to a bar at some point, but a coupon is unlikely to make someone go to a bar when they wouldn't
# otherwise.
```

### Independent Investigation

```python
# Lets take a look a coffee coupons
coffee_coupons = data[data["coupon"] == "Coffee House"]
# copy the dataset
coffee_coupons = coffee_coupons.copy()
coffee_coupons.head()
```

```python
# Get 3 or less acceptance rate
three_or_fewer = coffee_coupons[coffee_coupons['CoffeeHouse'].isin(['never', 'less1', '1~3'])]
three_or_fewer_acceptance_rate = three_or_fewer["Y"].mean()

# Get 4 or more acceptance rate
four_or_more = coffee_coupons[coffee_coupons['CoffeeHouse'].isin(['4~8', 'gt8'])]
four_or_more_acceptance_rate = four_or_more["Y"].mean()


# Create a bar plot comparing both values
plt.bar(["Three or Fewer", "Four or More"], [three_or_fewer_acceptance_rate, four_or_more_acceptance_rate])
plt.title('Coffe House Coupon Acceptance Rate by monthly visits')
plt.ylabel('Coupon Acceptance Rate')
plt.xlabel('Coffee House Visits per month')
plt.ylim(0, 1)  # Rates range between 0 and 1
plt.show()

# Again, we see that people who visit a coffee house more often are much more likely to accept
```

```python
# Lets investigate whether the time has a
acceptance_rate_by_time = coffee_coupons.groupby("time")["Y"].mean()
acceptance_rate_by_time.plot()

plt.title("Coffe House Acceptance Rate by Time")
plt.xlabel("Time")
plt.ylabel("Acceptance Rate")
plt.tight_layout()
plt.show()
# It appears time also has a large impact on whether a coupon is accepted or not. We can see that in the morning, and early
# afternoon, times when people would normally get a coffee, the acceptance rate is much higher.
```

```python
# First convert age to an int and set ages 50+ to 50 and under 21 to 20
def convert_age_to_int(age):
    if age == "50plus":
        return 50
    elif age == "below21":
        return 20
    return int(age)

# Apply conversion function
coffee_coupons['age'] = coffee_coupons['age'].apply(convert_age_to_int)

# go to bars more than once a month, had passengers that were not a kid, and were not widowed OR
group1 = coffee_coupons[(coffee_coupons['CoffeeHouse'].isin(['1~3', '4~8', 'gt8'])) & (~coffee_coupons["passanger"].isin(["Alone", "Kid(s)"])) & (coffee_coupons["maritalStatus"] != "Widowed")]
group1_acceptance_rate = group1["Y"].mean()
# print(group1)

# go to bars more than once a month and are under the age of 30 OR
group2 = coffee_coupons[(coffee_coupons['CoffeeHouse'].isin(['1~3', '4~8', 'gt8'])) & (coffee_coupons["age"] < 30)]
group2_acceptance_rate = group2["Y"].mean()
# print(group2)

# go to cheap restaurants more than 4 times a month and income is less than 50K.
group3 = coffee_coupons[(coffee_coupons["RestaurantLessThan20"].isin(['4~8', 'gt8'])) & (coffee_coupons["income"].isin(["Less than $12500", "$12500 - $24999", "$25000 - $37499", "$37500 - $49999"]))]
group3_acceptance_rate = group3["Y"].mean()

# All
all_acceptance_rate = coffee_coupons["Y"].mean()

# # Create a bar plot comparing all values
plt.bar(["Group 1", "Group 2", "Group 3", "All"], [group1_acceptance_rate, group2_acceptance_rate, group3_acceptance_rate, all_acceptance_rate])
plt.title('Coffe House Coupon Acceptance Rate by Group')
plt.ylabel('Coupon Acceptance Rate')
plt.ylim(0, 1)
plt.show()
```

```python
# From this data we can see that many factors will effect if a driver will accept a coffee house coupon. One of the strongest
# indicators is again whether the driver often drinks coffee or not. However, it is clear that a driver is much more likely to
# accept the coupon if the time of day corresponds to a time when a user would drink a coffee.
# This varies from bar coupon acceptance, where the time of day has much less of an impact on the acceptance rate. You can see
# this in the graph below.

acceptance_rate_by_time = bar_coupons.groupby("time")["Y"].mean()
acceptance_rate_by_time.plot()

plt.title("Bar Coupon Acceptance Rate by Time")
plt.xlabel("Time")
plt.ylabel("Acceptance Rate")
plt.tight_layout()
plt.show()
```
